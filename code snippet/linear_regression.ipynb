More examples and variations of Linear regression using scikit-learn can be found at https://scikit-learn.org/stable/modules/linear_model.html

```python
import numpy as np
import matplotlib.pyplot as plt

d = 100  # number of dimensions
n = 1000 # number of samples
X = np.random.normal(0,1,(n,d))
X[:,0]=1
beta = np.random.normal(0,1,(d,1))
beta[np.abs(beta)<.3]=0
y = X@beta
y.shape
```

```python
from sklearn.linear_model import LinearRegression
LinRegObj = LinearRegression(fit_intercept=False).fit(X, y)
betahat=LinRegObj.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(betahat.T)
```
